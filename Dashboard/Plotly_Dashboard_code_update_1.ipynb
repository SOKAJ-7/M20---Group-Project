```python
!pip install dash_bootstrap_components
import numpy as np
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Output, Input
import plotly.express as px
import dash_bootstrap_components as dbc

```

```python
# Code for Feature importance by Genre, data needs to be collected by ML algorithm before changing the graph 

names = np.array(['duration_ms', 'loudness', 'danceability', 'tempo', 'energy',
       'valence', 'instrumentalness', 'mode'])

values = [np.array([0.26426963, 0.1713882 , 0.15244108, 0.11391293, 0.09864651,
        0.09115961, 0.09070688, 0.01747517]),
 np.array([0.22678292, 0.1566791 , 0.1417102 , 0.13444943, 0.11909646,
        0.11840187, 0.0954191 , 0.00746091]),
 np.array([0.31674335, 0.17147713, 0.13127659, 0.11779682, 0.10147957,
        0.09153847, 0.06420517, 0.0054829 ]),
 np.array([0.25095957, 0.19786388, 0.19056586, 0.1262529 , 0.10846616,
        0.0648168 , 0.0606962 , 0.00037864]),
 np.array([0.24330939, 0.14303532, 0.13180711, 0.12130193, 0.11281486,
        0.11038832, 0.10222641, 0.03511666]),
 np.array([0.18420055, 0.16400739, 0.16302037, 0.15549666, 0.13914511,
        0.11085031, 0.06641271, 0.0168669 ])]

genres = ['rock', 'r&b', 'pop', 'edm', 'latin', 'rap']


figs = {
    f'{genres[i]}': px.pie(names = names, values = values[i], title = 'Feauture Importances by Genre', template = 'plotly').update_traces(name = genres[i], visible = False)
    for i in np.arange(0,6)
}

defaultgenre = genres[0]
fig = figs[defaultgenre].update_traces(visible=True)
for k in figs.keys():
    if k != defaultgenre:
        fig.add_traces(figs[k].data)
        
fig.update_layout(
    updatemenus=[
        {
            "buttons": [
                {
                    "label": k,
                    "method": "update",
                    # list comprehension for which traces are visible
                    "args": [{"visible": [kk == k for kk in figs.keys()]}],
                }
                for k in figs.keys()
            ]
        }
    ]
)
    
```

```python
# Code for Genre popularity violin plots filterable by genre
songs_df = pd.read_csv('spotify_merged.csv')
genre_list = ['pop', 'r&b', 'rap', 'latin', 'edm', 'hip_hop', 'rock', 'metal',
       'soul', 'motown', 'funk', 'folk', 'country', 'singer-songwriter',
       'blues', 'jazz', 'classical']

figures = {
    f'{genre_list[i]}':px.violin(songs_df['popularity'][songs_df['genre']==f'{genre_list[i]}'], template = 'plotly').update_traces(name=genre_list[i],visible = False)
    for i in np.arange(0,17)
}

defaultgenre = genre_list[0]
violin_pop = figures[defaultgenre].update_traces(visible=True)
for k in figures.keys():
    if k != defaultgenre:
        violin_pop.add_traces(figures[k].data)
        
violin_pop.update_layout(
    updatemenus=[
        {
            "buttons": [
                {
                    "label": k,
                    "method": "update",
                    # list comprehension for which traces are visible
                    "args": [{"visible": [kk == k for kk in figures.keys()]}],
                }
                for k in figures.keys()
            ]
        }
    ]
)
```

```python
app = dash.Dash(external_stylesheets = [dbc.themes.SLATE])
```

```python
# This is the html plus app.py equivalent, lots of documentation online to help find out syntax and how to add features. You need to run this script first and then run the 'Connect to Server' cell below
#If you need to make changes to the app layout after connecting to the server, press the square button below 'kernel', make changes to and run this cell, and then run the 'Connect to Server' cell


app.layout = html.Div(children=[
    # All elements from the top of the page
    html.Div([
        html.H1('Spotify Song Popularity Predictor'),

        html.Div(children='''
             This tool seeks to help artists in lesser known genres tailor their music to acheive greater mainstream appeal.
            
            Our algorithm processes the following characteristics of a song to predict its popularity:
                duration_ms, 
                danceability,
                energy,
                loudness, 
                mode, 
                instrumentalness,
                valence,
                tempo,
            
            The result is the song of your choice being determined as one of three classes: Unpopular, Somewhat Popular, or Popular. Furthermore, one can see which 
            features of a song provide the most impact in determining how popular a song is in any given genre.
        '''),
    ]),
     # New Div
    html.Div([
        dbc.Row(
            [
            (dbc.Col(dcc.Graph(
            id='',
            figure=fig),
                       width = {'size':4,'offset': 0,'order':1}
                       )),
               
         dbc.Col(dcc.Graph(
            id='violin',
            figure=violin_pop),
                       width = {'size':4,"offset": 0, 'order': 2}
                       ),
                
        dbc.Col(dcc.Graph(
            id='violi',
            figure=violin_pop),
                       width = {'size':4,"offset": 0,'order': 3}
                       )
            
               
        
            ])
    ]),
  

  
      
    
])
```

```python
# Connect to Server
if __name__ == '__main__':
    app.run_server(port = 4360)
```
